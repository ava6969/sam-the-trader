```python
# Minute data example POLYGON
import pytz
import sys
NY = 'America/New_York'
df_list = []
start = pd.Timestamp('2004-01-02 09:30', tz=NY).value/1e6
end = pd.Timestamp('2020-01-02 16:00', tz=NY).value/1e6
```

```python
from IPython.display import clear_output
while start < end:
    df = api.polygon.historic_agg_v2('AAPL', 1, 'minute', _from=start, to=end).df
    df_list.append(df)
    start = df.index[-1].value/1e6
    clear_output(wait=True)
    print(f'current date: {df.index[-1]}')
    
```

```python
new_df = pd.concat(df_list)
new_df
```
